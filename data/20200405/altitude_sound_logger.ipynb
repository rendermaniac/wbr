```python
import pandas as pd
alt3 = pd.read_csv("alt3.csv")
alt4 = pd.read_csv("alt4.csv")
```

```python
alt3.plot(y="Altitude", x="Time")
```

If we plot recorded altitude against Time (in milliseconds) we can see that the rocket took off at approximately 35 seconds, reached apogee of about 3 metres above ground level and gently descended on the parachute.

```python
alt3['Sound Normalized'] = (alt3.Sound - 580) / 100.0
alt3.plot(y=["Altitude", "Sound Normalized"], x="Time")
```

The sound plot shows that there are at least 2 distinct sound events. One at launch, and one just after apogee.
We could use these to detect launch and release the parachute. Unfortunately there is also some noise from presurising the bottle which could cause false alerts.

```python
alt4['Sound Normalized'] = (alt4.Sound - 580) / 100.0
alt3.plot(y=["Altitude", "Sound Normalized"], x="Time")
alt4.plot(y=["Altitude", "Sound Normalized"], x="Time")
```

These distinct spikes can be seen on the other flight where we successfully recorded data. This subsequent flight did not reach as high as the first and launched earlier.

Note that we don't really need to use this data to control recording, and recording a minute (60000) from when we start pressurising seems sufficient. This means that we can use our limited processing power for other tasks.
